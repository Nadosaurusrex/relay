## Policy Implementation

Here's the policy that prevents this scenario:

```yaml
policies:
  - name: "gdpr_data_protection"
    description: "Enforce GDPR data minimization and compliance"
    rules:
      - id: "limit_customer_data_queries"
        condition:
          provider: "database"
          method: "query_customers"
          parameter_constraints:
            limit:
              max: 100  # Maximum 100 records per query
            fields:
              # Only allow non-PII fields without approval
              allowed: ["user_id", "created_at", "subscription_tier"]
        action: "allow"
        reason: "Within GDPR data minimization limits"
      
      - id: "require_dpo_approval_bulk_export"
        condition:
          provider: "database"
          method: "query_customers"
        action: "deny"
        reason: "Bulk customer data export requires DPO approval"
        escalation:
          to: "data-protection-officer"
          include: ["request_details", "legal_basis", "purpose"]
      
      - id: "block_pii_to_third_parties"
        condition:
          provider: "database"
          method: "export_customers"
          destination:
            type: "external"
        action: "deny"
        reason: "PII export to third parties requires DPA and consent"
```

**GDPR Principles Enforced:**

1. **Data Minimization** (Art. 5.1c): Limit queries to minimum necessary
2. **Purpose Limitation** (Art. 5.1b): Enforce intended use
3. **Accountability** (Art. 5.2): Complete audit trails
4. **Security** (Art. 32): Technical measures to prevent breaches
